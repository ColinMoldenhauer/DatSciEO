```python
import os
import numpy as np
from sklearn.model_selection import cross_val_score, train_test_split
# I misunderstand it before. 
# The MLP used here is not completely equivalent to the BP neural network. 
# It just uses Backpropagation for optimization.
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
# Sometimes I run the classifier, this warning will occurs
# UndefinedMetricWarning: Precision and F-score are ill-defined and being set to 0.0 in labels with no predicted samples. 
#     Use `zero_division` parameter to control this behavior.
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.exceptions import ConvergenceWarning
```

```python
# Define the list of tree species
tree_species = ["Acer_pseudoplatanus-", "Betula_pendula-", "Carpinus_betulus-", "Fagus_sylvatica-", "Fraxinus_excelsior-",
                "Picea_abies_1123-", "Pinus_sylvestris-", "Quercus_petraea-", "Quercus_robur-", "Sorbus_aucuparia-"]
#Each species has different files numbers. In addition, some files are missing within a species. 
#For example, "Acer_pseudoplatanus_0", "Acer_pseudoplatanus_2","Acer_pseudoplatanus_3", ...
#We can't apply automatic search for end-file with while loop. Thus, I speficied the end num string below
tree_species_endnum = [745, 560, 881, 2276, 847,
                3760, 3166, 1168, 829, 243]
# Initialize empty lists to store data and labels
data = []
labels = []
# Specify data folder direction
data_dir = r'F:\Semester 1\DSEO\1123_delete_nan_samples_nanmean_B2'
# Check if the file exists before attempting to load it
for num, species in enumerate(tree_species):
    i = 0
    j = 0
    while True:
        file_path = os.path.join(data_dir, f"{species}{i}.npy")
        if os.path.exists(file_path):  # Check if the file exists
            try:
                file_data = np.load(file_path)
                data.append(file_data)
                labels.append(tree_species.index(species))
            except Exception as e:
                print(f"Error loading file: {file_path}")
                print(e)
        else:
            print(f"File not found: {file_path}")  # Print the path of non-existing files
        i += 1
        if i > tree_species_endnum[num]:
            break
```

```python
# Convert lists to NumPy arrays
data = np.array(data)
labels = np.array(labels)
```

```python
whether_green=np.zeros((data.shape[0], 1))
indices = np.where((labels == 5) | (labels == 6))
whether_green[indices] = 1
data_new = np.concatenate((data, whether_green), axis=1)
print (data_new.shape)
# The data is in dimension of (8182, 30)
# n_samples: 8128
# n_features: 30
```

```python
# Add a seed
np.random.seed(42)
# Split the data into test data and train data
data_train, data_test, labels_train, labels_test = train_test_split(data, labels, test_size = 0.3, random_state=42)

# Split the data into test data and train data_new
data_new_train, data_new_test, labels_new_train, labels_new_test = train_test_split(data_new, labels, test_size=0.3, random_state=42)
```

```python
# Bulid a MLP classifier
# hidden_layer_sizesarray-like of shape(n_layers - 2,), default=(100,)
# The default solver ‘adam’ works pretty well on relatively large datasets 
#    (with thousands of training samples or more) in terms of both training time and validation score. 
# Next step try to find the best size of hiden layer, e.g.add a Cross-validation
clf = MLPClassifier(activation='logistic',solver='adam',hidden_layer_sizes=(30,20), random_state=42)

# Cross Validation--K folds
# on train dataset
scores = cross_val_score(clf, data_train, labels_train, cv=5, scoring='accuracy')

# The score of every folds and its average
print("Cross-validation scores:", scores)
print("Average score:", scores.mean())

clf.fit(data_train, labels_train)
test_score = clf.score(data_test, labels_test)
print("Test set score:", test_score)

labels_hat = clf.predict(data_test)
print(classification_report(labels_test, labels_hat))
```

```python
mlp = MLPClassifier(activation='logistic',solver='adam',hidden_layer_sizes=(30,20), random_state=42)

# Cross Validation--K folds
# on train dataset
scores_new = cross_val_score(mlp, data_new_train, labels_new_train, cv=5, scoring='accuracy')

# The score of every folds and its average
print("Cross-validation scores:", scores_new)
print("Average score:", scores_new.mean())

mlp.fit(data_new_train, labels_new_train)
test_score_new = mlp.score(data_new_test, labels_new_test)
print("Test set score:", test_score_new)

labels_hat_new = mlp.predict(data_new_test)
print(classification_report(labels_new_test, labels_hat_new))
```
